<a id='1'><h2><font color = 'Maroon'>1. Problem Statement</h2></a>

Education is the precursor to a country’s potential growth. In the case of Portugal, it has seen a steady improvement in the overall education system over the last decades, albeit rather lagging behind its European counterparts. Putting it in perspective, the Portuguese education ranks at the rear end of Europe, with its high student failure and drop-out rates.

On the other hand, Singapore’s education system has been ranked amongst the best in the world. A 2018 article published by the Economist noted that Singapore students are 3 years ahead of their American peers and consistently top the OECD’s Programme for International Student Assessment (PISA) (a triennial test for 15-year-olds comprising three main categories of maths, reading and science)(1). Another article, published by the BBC in 2016, reported that Singapore tops the rankings for secondary pupils in international tests on math and science(2). The only western European countries to appear in the global top 10 were England, Norway, and Ireland. Portugal did not make the cut.

To investigate this shortcoming, the Portuguese Ministry of Education has appointed our data science team in Singapore to use their data and conduct exploratory data analysis (EDA) to answer the big question of what makes some students excel, and some others fail. To achieve this, we are going to look into the sample data from secondary students of two schools. These students are in their 3rd year of secondary school education, which is equivalent to 12th grade or pre-college stage in other countries.

The focus of our analysis is to uncover the formula behind high-performing students and to understand the unfortunate cause of the lower-performing ones. This means finding correlation between the students’ 33 attributes and their final grade of exams. These attributes range from the students’ fundamental classification such as sex, health, family wealth, to their family closeness and social proclivity such as their frequency of after-school hangout and alcohol consumption. 

We hope that these findings will better guide and inform the Portuguese Ministry of Education in crafting education policies that will see an improvement in the standards of their secondary school education.

<p> Looking through the 33 attributes that are available in the dataset, our purpose is to find correlations and understand the factors at play that contribute to why some students excel while some don't. From the attributes in the dataset we have identified several primary drivers of students' performance. These 4 drivers are specified as the following categories: 

- Category 1 : Social Proclivity
- Category 2 : Family Background and Intimacy
- Category 3 : Academic Inclination and Infrastructure
- Category 4 : Fundamental Demographic Classification


#### Category 1 : Social Proclivity
Attributes used : Dalc, Walc, goout, romantic, freetime

We would like to understand if a student's alcohol consumption, frequency of going out, romantic relationship and amount of free time after school affects the student's performance in school. Our hypothesis is that all the attributes have a negative relationship with the students' grades. We apply the k-means algorithm and analyse the proportion of students who failed in each cluster. We also build a decision tree to predict a students' grade (pass or fail) based on the attributes relating to social proclivity.
    
#### Category 2 : Family Background and Intimacy
Attributes used : famsize, famrel, Pstatus, guardian, Fedu, Medu, Mjob, Fjob

In this cateogry, we would like to find out whether family background has an effect such that we would be able to predict a student's grades based on qualities related to the student's family background. Our hypothesis is that the right combination of features related to this category could accurately predict it so. The chosen atttributes are split into 2 subsets pertaining to 2 themes: parents' background (i.e. education and jobs) and family closeness (i.e. quality of family relationships). Two decisions trees would be used to analyse the above 2 subsets. We will also find the singular most important features in each subset, which would give insight as to possible targets for any intervention programs by the school. Lastly, an unsupervised machine learning technique, k-means, would be used on the whole features set to see if further insights could be derived. 

#### Category 3 : Academic Inclination and Infrastructure
Attributes used : studytime, schoolsup, famsup, paid, activities, nursery, higher, internet, absences, traveltime

It is important to identify if academic inclination and environmental supports play a role in affecting students academic performance. We will analyse the amount of study time, willingness to take a higher education, travel time to school, absences rate ,etc to understand students pass fail rate. Thereafter, students will be classified based on clustering further enhance our understanding on their grades performances. Also we will predict whether a student will pass/ fail based on decision tree modelling.

Our hypothesis is that students who are more willing to study and put more effort to academics as well as provided with good surrounding supports will likely to perform better.

#### Category 4 : Fundamental and Demographic Classification
Attributes used : school, sex, age, address, famsize, health, reason, failures

In this category, we want to understand if students' performance are likely influenced by their fundamental/ innate attributes. We will take a closer look at the school they go to, which gender performs better, their residence type, etc. Students will be classified by their fundamental attributes using Decision Tree model for the purpose of understanding if their innate attributes contribute to their chances of then getting a passing grade.

Our hypothesis is that students who have fewer prior failures, better health, have a city address, and come from better school are more likely to do better. Although we have no pre-assumptions on whether which gender, age or residence better students belong to, the analysis and correlation will be uncovered later in the chapter.

### Now, we proceed to apply the k-means clustering algorithm and build a decision tree based on the social proclivity variables.

```python
df1 = all_por_df_encoded[['goout', 'Dalc', 'Walc', 'romantic', 'freetime']]
display(df1.head())
```

#### K-means clustering

```python
# plot elbow plot and select a k
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt

X = df1
distorsions = []
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    distorsions.append(kmeans.inertia_)

fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 20), distorsions)
plt.grid(True)
plt.xticks(ticks=range(2,20))
print(distorsions)
plt.title('Elbow curve') # select k = 10
```

#### Kmeans

Lastly, in this section, we will try tapping on unsupervised learning method Kmeans to see if a better model could be acheieved versus the Decision Tree. For this section, we will use all extracted features related to Category 2:

features = ["famsize", "famrel", "Pstatus", "guardian", "Mjob", "Fjob", "Medu", "Fedu"]

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
```

```python
# Elbow plot
X = all_por_df_encoded[["famsize", "famrel", "Pstatus", "guardian", 
                        "Mjob", "Fjob", "Medu", "Fedu"]]
distortions = []

for k in range(2,21):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    distortions.append(kmeans.inertia_)

fig = plt.figure(figsize=(15,5))
plt.plot(range(2,21), distortions)
plt.xticks(range(2,21))
plt.grid(True)
plt.show()
```

```python
# Fit model with chosen k
k = 2
model = KMeans(n_clusters=k, random_state=0)
family_cluster_kmeans = model.fit(X)

print(family_cluster_kmeans)
```

```python
# Predicting with kmeans
trained_family_cluster_labels = family_cluster_kmeans.predict(X)
```

```python
# Finding out silhouette_score for chosen k
silhouette_avg = silhouette_score(X, trained_family_cluster_labels)

print("{:.3f}".format(silhouette_avg))
```

```python
# Silhouette scores with other k
for k in [2,4,8,16]:
    
    kmeans_setting= KMeans(n_clusters=k, random_state=0)
    kmeans = kmeans_setting.fit(X)
    cluster_labels = kmeans.predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    
    print("For k = {}, silhouette score is {:.3f}.".format(k, silhouette_avg))
```

```python
# Getting labels
all_por_df_encoded_family = all_por_df_encoded.copy()
all_por_df_encoded_family["family_cluster_labels"] = family_cluster_kmeans.labels_

display(all_por_df_encoded_family.head())
```

```python
# Cluster centroids
family_cluster_centroids_df = pd.DataFrame(family_cluster_kmeans.cluster_centers_,
                                           columns=[X.columns[:8]]) 

family_cluster_centroids_df['label'] = family_cluster_centroids_df.index

display(family_cluster_centroids_df)
```

```python
# Matching G3_bins against labels
gb_labels = all_por_df_encoded_family.groupby(["family_cluster_labels", "G3_bins"]).size()
display(gb_labels)
```

#### Conclusion :
Kmeans on all extracted features related to Category 2 did not perform well.

Insights:
- Firstly, we note the overall poor silhouette score across different k-values. While it is not a negative score, it is very near to 0, which means the labels produced by the model are on or very close to the decision boundary between neighboring clusters. 
- Secondly, it was also noted that increasing the number of clusters caused the silhouette score to drop even further. Hence, increasing the number of clusters may actually have a counter-intuitive effect. For simplicity's sake, we went with k = 2 then.
- Lastly, as can be seen from the close numbers of labels in both "pass" and "fail" in G3_bins, there was a lot of overlap made by the kmeans classifier.

As such, we feel this model would not work very well.

```python
# Machine Learning - Clustering - Predict Student Grades by infrastructure

all_por_df_cluster_infrastucture = all_por_df_encoded[['traveltime', 'studytime', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet','absences']]

from sklearn.cluster import KMeans
from matplotlib import pyplot as plt

X = all_por_df_cluster_infrastucture
distorsions = []
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    distorsions.append(kmeans.inertia_)

fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 20), distorsions)
plt.grid(True)
plt.xticks(ticks = range (2,20))
plt.title('Elbow curve') # select k = 6
```

- On decision tree, model is able to predict positive true value which suggested that the accuracy is high (85%). However, it is not good in predicting true negative value. In this case, there is a need to find a balance to avoid overfitting or pruning by adjusting the max tree depth value

<a id='4.2'><h3>4.2. Limitations and Future Improvements</h3></a>

- It appears that discrete attributes contain quite a limited range of unique values. For example, Fjob and Mjob only comprises of 4 options, and anything other than the pre-given option is labelled as 'other'. This could be further improved by better survey method and data gathering
- 'Reason' attribute is a big missed opportunity on potential NLP analysis - Future suggestion is improvement on the data gathering method by allowing for text/descriptive inputs
- Limitations on this dataset is that both traveltime and studytime attributes are quite definite (i.e 1 or 2 or 3 or 4 hours). Therefore , it is hard to create a visualization on k-clustering. Also, polynomial regression can be explored or other more advanced machine learning models
- More insights can possibly be obtained from 'freetime' attritubtes if the students were asked what they do during their free time since they can be studying or relaxing
- On predictive modelling: for numeric attributes more regression models (such as polynomial regression) can be used to provide a better estimation of the final grade.
